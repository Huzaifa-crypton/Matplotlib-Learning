# Subplots plotting
We can plot subplots using two methods which are described below:

```python
# Option 1
fig , ((ax1 , ax2) , (ax3 , ax4)) = plt.subplots(nrows = 2 , ncols= 2 , )

ax1.plot(x , y)
ax1.set(title = "Simple Plot" , xlabel = "x axis" , ylabel = "y-axis")
ax2.scatter(x , y)
ax3.barh( x ,y)
ax4.bar(x , y)
```

```python
# Option 2
fig , (ax) = plt.subplots(nrows = 2 , ncols= 2 , )

ax[0][0].plot(x , y);
ax[0][0].set(title = "Simple Plot" , xlabel = "x axis" , ylabel = "y-axis");
ax[0][1].scatter(x , y);
ax[1][0].barh( x ,y);
ax[1][1].bar(x , y);
```

### Converting price to the correct format

```python
car_sales["Price"]=car_sales["Price"].replace('\D', '', regex=True)
car_sales["Price"] = car_sales["Price"].str[0:-2].astype(int)
car_sales
```

### Adding Date column to the car sales data frame

```python
car_sales["Dates"] = pd.Series(pd.date_range('1/1/2023' , periods = len(car_sales)))
car_sales
```

### Adding total sales column by applying the cummulative operation

```python
car_sales["Total Sales"] = pd.Series(car_sales["Price"].cumsum())
car_sales
```

### Plotting graph between total sales and dates

```python
car_sales.plot(x="Dates" , y="Total Sales" , kind = "scatter");
```

```python
car_sales.plot(x="Dates" , y="Total Sales" , kind = "line" , xlabel = "Dates" , ylabel = "Total Price");
```

# Different Dataset

```python
heart_disease = pd.read_csv("heart-disease.csv")
heart_disease
```

```python
heart_disease.plot.hist(subplots = True , figsize = (10,20));
```

```python
over50 = heart_disease[heart_disease['age'] > 50]
```

```python
over50.head()
```

```python
over50['age'].plot(kind = "hist" , bins = 70)
```

```python
over50.plot( c = 'target' , x="age" , y="chol" , kind = 'scatter' ,figsize = (10,6));
```

# Object Oriented Method + PyPlot

```python
fig , ax = plt.subplots(figsize=(10,6))
over50.plot( c = 'target' , x="age" , y="chol" , kind = 'scatter' ,ax=ax);
ax.set_xlim(40,100)
```

## OO method from scratch

```python
fig , ax = plt.subplots(figsize= (10,6))

#Plot the graph
scatter = ax.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target']);

# Customize figure
ax.set(title = "Heart disease and cholestrol",
    xlabel = 'age',
    ylabel = 'cholestrol');

# Set legend of the figure
ax.legend(*scatter.legend_elements() , title = "Target");

# Add horizontal line
ax.axhline(over50['chol'].mean() , linestyle = "--")
```

# Subplotting cholestrol, age, and thalac

```python
fig , (ax0 , ax1) = plt.subplots(nrows = 2 , ncols= 1 ,figsize = (10,10), sharex = True);

# Working on axis 0
scatter = ax0.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target']);

ax0.set(title = "Heart disease and cholestrol level", ylabel = 'cholestrol')

ax0.legend(*scatter.legend_elements() , title = "Title");

ax0.axhline(y = over50['chol'].mean() , linestyle = '--');

# Working on axis 1
scatter1 = ax1.scatter(x=over50['age'],
                     y=over50['thalach'],
                     c=over50['target']);

ax1.set(title = "Heart disease and thalac level" , xlabel = 'Age' , ylabel = 'Thalac')

ax1.legend(*scatter1.legend_elements() , title = "Title");

ax1.axhline(y = over50['thalach'].mean() , linestyle = '--');

# Set figure title
fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight = "bold" )
```

# Final graph with styles and customization
#### cmap helps in adding theme

```python
plt.style.use('seaborn-v0_8-whitegrid')
fig , (ax0 , ax1) = plt.subplots(nrows = 2 , ncols= 1 ,figsize = (10,10), sharex = True);

# Working on axis 0
scatter = ax0.scatter(x=over50['age'],
                     y=over50['chol'],
                     c=over50['target'] ,
                     cmap = 'winter');

ax0.set_xlim([50 , 80])
ax0.set_ylim([40 , 600])

ax0.set(title = "Heart disease and cholestrol level", ylabel = 'cholestrol')

ax0.legend(*scatter.legend_elements() , title = "Title");

ax0.axhline(y = over50['chol'].mean() , linestyle = '--' , color = 'red');

# Working on axis 1
scatter1 = ax1.scatter(x=over50['age'],
                     y=over50['thalach'],
                     c=over50['target'],
                      cmap = 'winter');

ax1.set_ylim([60 , 200])

ax1.set(title = "Heart disease and thalac level" , xlabel = 'Age' , ylabel = 'Thalac')

ax1.legend(*scatter1.legend_elements() , title = "Title");

ax1.axhline(y = over50['thalach'].mean() , linestyle = '--' , color = "red");

# Set figure title
fig.suptitle("Heart Disease Analysis" , fontsize = 16 , fontweight = "bold" )
```
